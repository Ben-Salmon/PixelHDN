```python
# We import all our dependencies.
import numpy as np
import torch
from torchvision import transforms
from tifffile import imread, imsave
import sys
sys.path.append('../')
from HDN.models.lvae import LadderVAE
from utils import predict_and_save
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
observation= torch.from_numpy(imread('../data/Convallaria/observation.tif')).type(torch.float)
signal = torch.mean(observation, axis=0)

num_obs = 1

observation_cropped = torch.zeros(64*num_obs,1,128,128)
signal_cropped = torch.zeros_like(observation_cropped)

counter = 0
for i in range(num_obs):
    for j in range(8):
        for k in range(8):
            observation_cropped[counter] = observation[i, 0, (j*128):((j+1)*128), (k*128):((k+1)*128)]
            signal_cropped[counter] = signal[0, (j*128):((j+1)*128), (k*128):((k+1)*128)]
            counter += 1
```

```python
HDN = LadderVAE.load_from_checkpoint('../dn_checkpoint/Convallaria/signal_dependent/HDN/checkpoints/epoch=7899-step=78999.ckpt').to(device)

predict_and_save.predict_and_save(HDN, observation_cropped, '../results/Convallaria/signal_dependent', 100)

imsave('../results/Convallaria/signal.tif', signal_cropped.numpy())
```

```python
def PSNR(gt, img, psnrRange):
    mse = np.mean(np.square(gt - img))
    return 20 * np.log10(psnrRange) - 10 * np.log10(mse)

result = imread('../results/Convallaria/signal_dependent/mmse.tif')
signal = imread('../results/Convallaria/signal.tif')
samples = imread('../results/Convallaria/signal_dependent/samples.tif')

psnr = np.array([])
for i in range(result.shape[0]):
    range_psnr = np.max(signal[i])-np.min(signal[i])
    psnr = np.append(psnr, PSNR(signal[i], result[i], range_psnr))

print(np.mean(psnr))
```
